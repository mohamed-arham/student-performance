```python
import pandas as pd
from google.colab import files

# Upload the CSV manually
uploaded = files.upload()

# Now load it (this grabs the uploaded file name automatically)
df = pd.read_csv(list(uploaded.keys())[0])
df.head()
```

```python
# Structure and stats
df.info()
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Gender-wise average scores
df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()

# Lunch-wise average scores
df.groupby('lunch')[['math score', 'reading score', 'writing score']].mean()

# Test preparation-wise scores
df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Math Score by Gender
sns.boxplot(x='gender', y='math score', data=df)
plt.title('Math Score by Gender')
plt.show()

# Reading Score Distribution
sns.histplot(df['reading score'], kde=True, bins=20)
plt.title('Reading Score Distribution')
plt.show()

# Writing Score by Test Preparation
sns.barplot(x='test preparation course', y='writing score', data=df)
plt.title('Writing Score vs Test Preparation')
plt.show()
```
